```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 100 observations from a normal distribution with mean=2 and std=0.5
data = np.random.normal(loc=2, scale=0.5, size=100)

# Plot histogram of the observations
fig, ax = plt.subplots(figsize=(24, 16))
ax.hist(data, bins=20, color='skyblue', edgecolor='black')
ax.set_title('Histogram of Observations from Normal Distribution', fontsize=20)
ax.set_xlabel('Value', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)


```

```python
# Create cumulative sum of the observations
cumulative_sum = np.cumsum(data)

# Plot cumulative sum
fig, ax = plt.subplots(figsize=(24, 16))
ax.plot(cumulative_sum, marker='o', linestyle='-', color='purple')
ax.set_title('Cumulative Sum of Observations', fontsize=20)
ax.set_xlabel('Observation Number', fontsize=16)
ax.set_ylabel('Cumulative Sum', fontsize=16)
plt.grid(True)
```
