```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#1 Create an array with (100) elements using np.arange()
array_100 = np.arange(100)
print(array_100)
```

```python
#2 Create two numpy arrays whose values are neither all 0 or 1. 
# Multiply the two arrays and save the result i.e., result = array1 * array2.
array1 = np.random.rand(10) * 10
array2 = np.random.rand(10)
result = array1 * array2
print(result)
```

```python
#3 Create a dictionary where each key points to a unique numpy array. 
# All arrays should be of the same length. Then transform that dictionary into a pandas dataframe.

array_dict = {
    "array1": np.random.rand(10),
    "array2": np.random.rand(10)*10,
    "array3": np.random.rand(10)*100
}
print(array_dict["array2"])
df = pd.DataFrame(array_dict)
print(df)
```

```python
#4 Import a data set for your project in pandas. 
# You may use whatever means you prefer to import the data. Print the dataframe.
iris_df = pd.read_csv("Iris.csv")
iris_df.head(7)
```

```python
#5 Create a new column of data derived from the data that you have already imported. 
# hint: this may be a ratio or whatever transformation that you think would be useful...

iris_df['petal_sepal_length_ratio'] = iris_df['PetalLengthCm'] / iris_df['SepalLengthCm']
iris_df.head(7)
```

```python
iris_df.info()
```

```python
#6 Use a for loop to print individually each column in the dataframe.
for col in iris_df:
    print(f"Column: {col}")
    print(iris_df[col].head(), "\n") 
```

```python
# Create a function that plots 2 to 3 different variables from the dataframe. 
# The function may look something like "def plot_data(df,names, etc...). 
# Each plot should include a legend and title.

def plot_data(df, column_names):
    if len(column_names) < 2 or len(column_names) > 3:
        raise ValueError("Please provide 2 or 3 column names for plotting")

    df[column_names].plot(kind="line")
    plt.title(f"Line Plot of { column_names }")
    plt.xlabel("Index")
    plt.ylabel("Values")
    plt.legend(column_names)
    plt.show()


plot_data(iris_df, ['PetalLengthCm', 'SepalLengthCm'] )
```

```python
#8 Create a bar graph plot using the data you have imported.
species_counts = iris_df['Species'].value_counts()
species_counts.plot(kind='bar', color='skyblue')
plt.title("Bar Graph of Species Counts")
plt.xlabel("Species")
plt.ylabel("Counts")
plt.show()
```

```python
#9 On any of the graphs you have created, change the font size, axis labels, tick labels, color, linewidth, and so forth. 
# Adjust these parameters to make the visualization aesthetically appealing..

iris_df[['SepalWidthCm', 'PetalWidthCm']].plot(kind='line', color=['red', 'green'], linewidth=2)
plt.title("Customized Line Plot of SepalWidthCm and PetalWidthCm", fontsize=14)
plt.xlabel("Index", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(['SepalWidthCm', 'PetalWidthCm'], fontsize=10)
plt.grid(True)
plt.show()
```
