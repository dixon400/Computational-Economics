6. Explain why it might be advantageous to create a function instead of building all commands from scratch as you create a script.


Developing functions in programming provides notable benefits that improve the effectiveness and excellence of code development. Functions facilitate code reusability by encapsulating certain tasks into reusable blocks of code, enabling programmers to utilise the same code in various sections of a project or in separate projects. This not only saves time and effort but also guarantees uniformity in work completion. Functions enhance the modularity of code by dividing intricate scripts into smaller, more manageable sections. Modularity simplifies the comprehension, creation, examination, and troubleshooting of code by allowing each function to concentrate on a specific task and be managed separately.

Functions offer a significant advantage by enhancing the maintainability and readability of code. Updates or alterations can be efficiently implemented in a centralised area, minimising the likelihood of errors and discrepancies throughout the codebase. Descriptively named functions enhance script readability by clearly conveying the purpose of a code block, therefore facilitating documentation and increasing the codebase's accessibility to new developers or collaborators. Functions provide abstraction, which conceals the intricacies of individual tasks from the user, enabling them to concentrate on more advanced logic and design.

Functions provide code flexibility by allowing parameterization, which enables the execution of operations on multiple data sets by giving different arguments. This adaptability minimises redundant code and broadens the use of functions in various situations. Functions in collaborative settings enable a division of labour, allowing numerous developers to work on separate sections of a script simultaneously without disrupting each other's progress. This can greatly accelerate development times and enhance the overall quality of the code.

```python
import pandas as pd
import plotly.graph_objects as go

def gather_statistics_plotly(df, sample=True, round_to=3):
    # Create an entry in the dictionary for each column of data
    stats_dict = {key: {} for key in df}

    for key, vals in df.items():
        vals = vals['list']
        # Drop any row with a null value
        #vals = vals.dropna(axis=0)
        stats_dict[key]["mean"] = round(stats.mean(vals), round_to)
        stats_dict[key]["median"] = round(stats.median(vals), round_to)
        stats_dict[key]["variance"] = round(stats.variance(vals, sample), round_to)
        stats_dict[key]["SD"] = round(stats.SD(vals, sample), round_to)
        stats_dict[key]["SE"] = round(stats.STE(vals, sample), round_to)
        stats_dict[key]["skewness"] = round(stats.skewness(vals, sample), round_to)
        stats_dict[key]["kurtosis"] = round(stats.kurtosis(vals, sample), round_to)
    
    # Convert the stats dictionary to a DataFrame for easier plotting
    stats_df = pd.DataFrame(stats_dict).T
    
    # Plotting
    fig = go.Figure()
    for column in stats_df.columns:
        fig.add_trace(go.Bar(name=column, x=stats_df.index, y=stats_df[column]))
    
    fig.update_layout(title="Statistical Measures of DataFrame Columns",
                      xaxis_title="Columns",
                      yaxis_title="Value",
                      barmode='group')
    fig.show()
```

```python
new_nested_dict = nested_dict

for x in range(1, 11):
    del new_nested_dict[x]['variance']
    
#print(new_nested_dict)
```

```python
gather_statistics_plotly(new_nested_dict)
```
