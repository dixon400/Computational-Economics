1. Generate data for one random variable with a mean of 5 and standard deviation of 1. Run a one-sample t-test that the mean of the sample is different than 0. Perform the same test again with a mean of 1 and a standard deviation of 1.

2. Why has the result changed? In answering this question, plot both samples with the index as the x-axis values. Use two colors to differentiate the samples. Indicate the value of zero using a horizontal line. Reference the plot in your answer.

3. Now run a t-test for the two samples. Print the results and explain their meaning.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Generate data for one random variable with mean of 5 and std dev of 1
np.random.seed(42)  # for reproducibility
sample1 = np.random.normal(loc=5, scale=1, size=100)

print(sample1)
# Perform a one-sample t-test that the mean of the sample is different than 0
t_test_result_1 = stats.ttest_1samp(sample1, popmean=0)
print(t_test_result_1)
```

```python
# Generate data for another random variable with mean of 1 and std dev of 1
sample2 = np.random.normal(loc=1, scale=1, size=100)
print(sample2)

# Perform the same test with mean of 1
t_test_result_2 = stats.ttest_1samp(sample2, popmean=0)
print(t_test_result_2)

# Plot both samples
plt.figure(figsize=(10, 6))
plt.plot(sample1, color='blue', label='Sample Mean=5, Std=1')
plt.plot(sample2, color='red', label='Sample Mean=1, Std=1')
plt.axhline(y=0, color='k', linestyle='--', label='Mean=0')
plt.legend()
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Samples with Different Means')
plt.show()
```

The results of the one-sample t-tests for the samples with means of 5 and 1 against a population mean of 0 are significantly different:

For the sample with a mean of 5 and a standard deviation of 1, the t-test statistic is approximately 53.91, with a p-value near 0 (3.44e-75), strongly rejecting the null hypothesis that its mean is equal to 0.
For the sample with a mean of 1 and a standard deviation of 1, the t-test statistic is approximately 10.72, with a p-value also near 0 (2.96e-18), similarly rejecting the null hypothesis but with a smaller t-statistic compared to the first sample.

The reason for the difference in the results is evident from the plot. The sample with a mean of 5 is much further away from 0, as indicated by its values predominantly lying above the horizontal line at 0, compared to the sample with a mean of 1, which is closer to the line, indicating a smaller distance from the hypothesized mean of 0. This is reflected in the higher t-statistic and the lower p-value for the sample with a mean of 5, suggesting a stronger rejection of the null hypothesis.

The independent t-test comparing the two samples resulted in a t-statistic of approximately 29.42 and a p-value near 0 (3.37e-74), strongly indicating that there is a significant difference in the means of the two samples. This means that we can reject the null hypothesis that the means of the two samples are equal, confirming that the observed differences in sample means (5 versus 1) are statistically significant.

```python
# Now, run a t-test for the two samples
t_test_result_3 = stats.ttest_ind(sample1, sample2)

t_test_result_1, t_test_result_2, t_test_result_3
```

The results provided from the t-tests reveal insights into the statistical significance and differences between two samples, as well as each sample's difference from a theoretical mean of 0. Here's a breakdown of what each result indicates:

1. **First One-Sample t-Test (Sample with Mean ~5)**:
    - **Statistic**: 53.91239478943712
    - **p-value**: 3.442588839645403e-75
    - **Degrees of Freedom (df)**: 99

   This result comes from testing whether the sample mean significantly differs from 0. The extremely high t-statistic combined with an exceedingly small p-value indicates a strong rejection of the null hypothesis. It suggests that the mean of this sample is significantly greater than 0. Given the sample size (100, inferred from df = n - 1), this result is highly reliable.

2. **Second One-Sample t-Test (Sample with Mean ~1)**:
    - **Statistic**: 10.719700685464499
    - **p-value**: 2.9623364942547248e-18
    - **Degrees of Freedom (df)**: 99

   This test also assesses if the sample mean is significantly different from 0. The lower t-statistic compared to the first test, but still significantly high, along with a very low p-value, indicates that the sample mean is significantly greater than 0, albeit the magnitude of difference is smaller compared to the first sample. This again is a robust result due to the large sample size.

3. **Independent Two-Sample t-Test (Comparing Sample1 and Sample2)**:
    - **Statistic**: 29.416180054185716
    - **p-value**: 3.3710475694504264e-74

   This test compares the means of the two samples directly. The large t-statistic and the minuscule p-value strongly suggest that the means of the two samples are significantly different from each other. The test doesn't report degrees of freedom, but for independent samples of equal size (100 each), it's effectively about assessing differences across the two groups as whole entities.

**Interpretation**: All tests confirm with strong statistical evidence that the means of both samples are significantly different from 0, and additionally, that the means of the two samples are significantly different from each other. The tests' p-values, being exceptionally close to 0, virtually eliminate the chance of these results occurring under the null hypotheses (no difference), indicating very strong evidence against the null in each case.
