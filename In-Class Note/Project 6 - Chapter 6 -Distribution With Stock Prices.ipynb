# Probability Distributions

## Types of Probability Distributions

### Normal Distributions

it describes a symmetric, bell-shaped curve characterized by its mean (μ) and standard deviation (σ). It models a wide range of natural phenomena and measurement errors.

<h3 align="center">
    <font size="4">
        $ X = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2} $
    </font>
</h3> 

If one knew the mean and standard deviation of a particular distribution, it would be possible to estimate the probability of drawing a value from a particular range - e.g., $a \leq X \leq b$ - by taking the definite integral of this function: 
<h3 align="center">
    <font size="4">
        $\int_{a}^{b}\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}dx$.
    </font>
</h3>

### Standard Normal Distribution

The standardized normal distribution is the most important member of the family of normal probability distributions—the one with $\mu = 0$  and $\sigma = 1$. The normal random variable distributed according to the standard normal distribution is called the standard normal variable and is denoted by $Z$. It represents the number of standard deviations a given point X lies from the mean of the standard normal distribution.  It is expressed as
<h3 align="center">
    <font size="4">
        $ Z =  \frac{X-\mu}{\sigma}$
    </h3>

The function `normal_dist` is designed to calculate the probability density of a normal (Gaussian) distribution given a mean (`mean`), standard deviation (`sd`), and an array of values (`x`). The normal distribution is one of the most important probability distributions in statistics, characterized by its bell-shaped curve. The probability density function (PDF) for a normal distribution is defined as:

\[ f(x | \mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} \]

where:
- \(x\) is the point at which to evaluate the function,
- \(\mu\) is the mean of the distribution,
- \(\sigma\) is the standard deviation of the distribution.

The function essentially implements this formula, where:
- `mean` corresponds to \(\mu\),
- `sd` corresponds to \(\sigma\),
- `x` is a NumPy array of values over which the PDF is evaluated. By default, `x` spans from -10 to 10 with 1000 points, providing a broad range to visualize the distribution.

```python
# Optionally, pass different array of x values tp be transformed; default is np.linspace(-5, 5, 1000)
def normal_dist(mean , sd, x = np.linspace(-10, 10, 1000)):
    prob_density = 1 / (sd * (2 * np.pi) ** .5) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

## Lognormal Distribution

Unlike the normal distribution, which is defined for any real number, the log-normal distribution is defined only for positive real numbers. It's a continuous probability distribution of a random variable whose logarithm is normally distributed. This means that if \(Y\) is a random variable with a normal distribution, then \(e^Y\) is log-normally distributed. The PDF of a log-normal distribution is given by:

\[ f(x) = \frac{1}{x\sigma\sqrt{2\pi}} e^{- \frac{(\ln x - \mu)^2}{2\sigma^2}} \]

where:
- \(x > 0\) is the value for which we're calculating the probability density,
- \(\mu\) and \(\sigma\) are the mean and standard deviation of the variable's natural logarithm, not the original variable itself.

This distribution is applicable in various areas, especially in finance for modeling stock prices, as well as in other fields where quantities grow exponentially and variability increases over time.

```python
import warnings
warnings.filterwarnings("ignore")
# Optionally, pass different array of x values tp be transformed; default is np.linspace(-5, 5, 1000)
def create_lognormal(mean, sd, x = np.linspace(0, 10, 1000)):
    density_points = (1 / (x * sd * np.sqrt(2 * np.pi))) * np.exp(-((np.log(x) - mean) ** 2) / (2 * sd ** 2))
    return density_points

fig, ax = plt.subplots(figsize=(10, 7))
x = np.linspace(0, 10, 1000)
lognorm = create_lognormal(0, 1,x)
ax.plot(x, lognorm)
plt.xscale("log")
ax.set_title("Lognormal Distribution with μ = 0 and σ = 1")
```

```python
def graph_yearly_distributions(data, key):
    years = sorted(list(set(data.index.year)))
    fig, ax = plt.subplots(figsize = (24, 16))
    for year in years:
        color = "C" + str(year % 12)
        year = str(year)
        plot_data = data[key].loc[year]
        plot_data.hist(bins = 20, color = color,
                       label = year, 
                       alpha = .6, 
                       ax = ax)
        ax.axvline(plot_data.mean(), 
                   color = color, 
                   linewidth = 5, 
                   ls = "--")
    ax.set_title(key)
    ax.legend(bbox_to_anchor = (1,1))
graph_yearly_distributions(log_data, "AAPL");
```
